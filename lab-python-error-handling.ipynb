# Step 1: Define the function for initializing the inventory with error handling
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory

````python
# 2  Modify the `calculate_total_price` function to include error handling.

def initialize_inventory(products):
    inventory = {}
    
    for product in products:
        while True:
            quantity = input("Enter the quantity of available " + product + "s : ")
            
            if quantity.isdigit():
                inventory[product] = int(quantity)
                break  
            else:
                print("Invalid input, please try again")
                
    return inventory
````

````python
# 3

def get_customer_order(inventory):
    while True:
        total_input = input("Enter the number of products you want to order:").strip()
        
        if total_input.isdigit():
            number_of_prod = int(total_input)
            if number_of_prod > 0:
                break 
            else:
                print("To place an order please introduce a number of products greater than 0.")
        else:
            print("Invalid input. Please enter the amount in the form of a whole number.")

    customer_order = []
    
   
    while len(customer_order) < number_of_prod:
        product = input("Enter the name of the product you want to order: ").strip().lower()
        
        
        if product not in inventory or inventory[product] <= 0:
            
            print("Please enter a valid product with stock available:" + str(list(inventory.items())))
        
        
        elif product in customer_order:
            print("You already added this product to your order.")
            
        else:
            customer_order.append(product)
            
    return customer_order
````

````python
def calculate_price(customer_order):
    prices = []
    for product in customer_order:
        while True:
            price_input = input("Enter the price of this product " + product + ": ").strip()
            
            if price_input.replace(".", "", 1).isdigit():
                price = float(price_input)
                if price > 0:
                    prices.append(price)
                    break 
                else:
                    print("The price of the product must be greater than 0.")
            else:
                print("Invalid input. Please enter a number and use '.' for decimals.")
                 
    total_price = sum(prices)
    return total_price




def calculate_order_statistics(customer_order, products):
    total = len(customer_order)
    percentage = (total / len(products)) * 100
    
    print("Order Statistics:")
    print("Total Products Ordered: " + str(total))
    print("Percentage of Unique Products Ordered: " + str(percentage))
    
    return total, percentage


def update_inventory(customer_order, inventory):
    updated_inventory = {}
    
    for product in inventory:
        value = inventory[product]
        
        if product in customer_order:
            new_value = value - 1
        else:
            new_value = value
            
        if new_value > 0:
            updated_inventory[product] = new_value
            
    return updated_inventory


def print_updated_inventory(inventory):
    print("Updated inventory:")
    
    final_inventory = []
    for product in inventory:
        quantity = inventory[product]
        print(product + ": " + str(quantity))
        
    return final_inventory
````

````python
products=['t-shirt','mug','hat','book','keychain']

print("--- STEP 1: INITIALIZING INVENTORY ---")
current_inventory = initialize_inventory(products)
print("Initial Inventory Dictionary:", current_inventory)
print("-" * 40)

print("\n--- STEP 2: GETTING CUSTOMER ORDER ---")
order = get_customer_order(current_inventory)
print("Customer Order List:", order)
print("-" * 40)

print("\n--- STEP 3: CALCULATING TOTAL PRICE ---")
total_to_pay = calculate_price(order)
print("Total Price to Pay: " + str(total_to_pay))
print("-" * 40)

print("\n--- STEP 4: SHOWING ORDER STATISTICS ---")
stats = calculate_order_statistics(order, products)
print("-" * 40)

print("\n--- STEP 5: UPDATING INVENTORY STOCK ---")
current_inventory = update_inventory(order, current_inventory)
print("-" * 40)

print("\n--- STEP 6: PRINTING THE FINAL INVENTORY ---")
print_updated_inventory(current_inventory)
print("=========================================")
````
